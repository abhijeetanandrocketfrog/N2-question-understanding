## Orien Questions

```python
import pandas as pd
import re
```

```python
def extract_eb_codes(row):
    pattern = r'@?(EB\d+)'
    eb_codes = re.findall(pattern, row)
    return ", ".join(eb_codes)

def extract_phrases(row):
    pattern = r'":\s*"([^"]+)"'
    phrases = re.findall(pattern, row)
    return ", ".join(phrases)
```

```python
file_name = "User_Questions_Orien_QA_2.xlsx"
data = pd.read_excel(file_name, sheet_name=1)
data = data[["Question_id", "Question", "Question Combination Keys (EB code → sub-code)"]]
data = data.dropna()
data
```

```python
# pattern = r'@?(EB\d+)'
# data = data[data["Question Combination Keys (EB code → sub-code)"].str.contains(pattern)]
# data
```

```python
data = data.copy()

data["EB Codes"] = data["Question Combination Keys (EB code → sub-code)"].apply(extract_eb_codes)
data["Phrases"] = data["Question Combination Keys (EB code → sub-code)"].apply(extract_phrases)
data
```

## Old Questions

```python
import json
import pandas as pd
import re
```

```python
with open("EB01-03-12.json", "r") as f:
    embedding = json.load(f)

def extract_eb_codes(row):
    EB_pattern = r"\bEB\d{2}\b"
    l = re.findall(EB_pattern, row)
    l2 = set()
    for term in l:
        if term == "EB01" or term == "EB03" or term == "EB12":
            l2.add(term)
    return ", ".join(l2)

def extract_phrases(row):
    l = []
    t2 = row.split(",")
    for code in t2:
        code = code.split()
        try:
            l.append(embedding[code[0]]["definition"])
        except:
            continue
    return ", ".join(l)
```

```python
file_name = "Entity Resolution User Questions_QA.xlsx"
data = pd.read_excel(file_name, header=1)
data = data[["Question", "cleaned_entities", "Entities_type [code]"]]
data
```

```python
data["EB Codes"] = data["Entities_type [code]"].apply(extract_eb_codes)
data["Phrases"] = data["Entities_type [code]"].apply(extract_phrases)
data
```

```python
# data = data.drop(columns=["Entities_type [code]", "cleaned_entities"])
data = data[data['Phrases'].str.len() > 0]
data
```

```python
data.to_excel("ground_truth_22.xlsx", index=False)
```
