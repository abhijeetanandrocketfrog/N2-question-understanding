```python
# Function to replace occurrences case-insensitively
def replace_clubbed_terms(text, mapping):
    if not isinstance(text, str):
        return text
    
    # Sort keys by length (longest first) so longer phrases get replaced first
    sorted_keys = sorted(mapping.keys(), key=len, reverse=True)
    
    for key in sorted_keys:
        pattern = re.escape(key)
        replacement = mapping[key]
        text = re.sub(pattern, replacement, text, flags=re.IGNORECASE)
    
    return text

# Apply to your dataframe
df["Phrases"] = df["Phrases"].apply(lambda x: replace_clubbed_terms(x, replacement_map))
df
```
